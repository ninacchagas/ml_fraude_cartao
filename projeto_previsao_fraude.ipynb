```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
```

```python
# importando o arquivo
df_credit = pd.read_csv('creditcard.csv', names=["Time","V1","V2","V3","V4","V5","V6","V7","V8","V9","V10","V11","V12","V13","V14","V15","V16","V17","V18","V19","V20","V21","V22","V23","V24","V25","V26","V27","V28","Amount","Class"])
df_credit
```

```python
# verificando os tipos dos dados não fraude]
df_nao_fraude = df_credit.Amount[df_credit.Class == 0]
df_nao_fraude
```

```python
# verificando os tipos dos dados de fraude
df_fraude = df_credit.Amount[df_credit.Class == 1]
df_fraude
```

```python
# distribuição dos dados tipo fraude e não fraude
df_credit.Class.value_counts()
```

```python
# verificando os tipos dos dados de fraude
df_fraude = df_credit[df_credit.Class == 1]
df_fraude
```

```python
# verificando os tipos dos dados de não fraude
df_nao_fraude = df_credit[df_credit.Class == 0]
df_nao_fraude
```

```python
# criando dataset com a mesma quantidade de fraude e não fraude
df_nao_fraudes = df_nao_fraude.sample(n=358)
df_nao_fraudes
```

```python
# concatenando dataframes de fraude e não fraude
df = pd.concat([df_nao_fraudes, df_fraude], axis=0)
df
```

```python
# ajustando o index
df.reset_index(inplace = True)
```

```python
df
```

```python
# Faremos um split, para pegar os dados de treinamento e teste e dados de validação (registros que o modelo nunca viu)
# retirar dados para validação
df_val_nao_fraude = df.head(5)
df_val_nao_fraude
```

```python
df_val_fraude = df.tail(5)
df_val_fraude
```

```python
# retirar os dados de validação do df
df = df.iloc[5:]

df = df.iloc[:-5]
df
```

```python
#concatenando df de validação
df_val = pd.concat([df_val_fraude, df_val_nao_fraude])
df_val.reset_index(inplace=True)
df_val
```

```python
df_val_total = df_val.Class
df_val = df_val.drop(['level_0','index','Time','Class'], axis=1)
```

```python
df_val
```

```python
# verificando a distribuição dos dados do tipo fraude e não fraude
df.Class.value_counts()
```

```python
# separando labels e features
X = df.drop(['index', 'Time', 'Class'], axis=1)
Y = df['Class']
```

```python
print(X,Y)
```

```python
# separando entre dados de treino e dados de teste
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)
```

```python
# criando o treinamento
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, Y_train)
pred = lr.predict(X_test)
acc = accuracy_score(Y_test, pred)

f'Acurácia: {acc * 100:.2f}'
```

```python
# validando
pred = lr.predict(df_val)
df = pd.DataFrame({'real': df_val_total, 'previsão': pred})

print(df)
```
